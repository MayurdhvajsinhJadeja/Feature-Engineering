```python
import numpy as np
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lr = Q1 - (1.5 * IQR)
    ur = Q3 + (1.5 * IQR)
    return lr, ur

lessVis, moreVis = remove_outlier(df['WAGE_RATE_OF_PAY_FROM_1'])
df['WAGE_RATE_OF_PAY_FROM_1'] = np.where(df['WAGE_RATE_OF_PAY_FROM_1']>moreVis,moreVis,df['WAGE_RATE_OF_PAY_FROM_1'])
df['WAGE_RATE_OF_PAY_FROM_1'] = np.where(df['WAGE_RATE_OF_PAY_FROM_1']<lessVis,lessVis,df['WAGE_RATE_OF_PAY_FROM_1'])

sns.boxplot(df['WAGE_RATE_OF_PAY_FROM_1'])
```
